### Explanation
displays Q1, Q2, Q3, Q4

box is Q2,Q4

Inter Quartile Range IQR = Q3-Q1

As a rule of thumb anything above Q4 + (IQR*1.5) is considered outlier

               anything below Q1 - (IQR*1.5) is considered outlier

## Line Plot

```python
x = np.arange(0,11)
y = x*2
plt.plot(x,y,color='g', linewidth=2, linestyle=":")
plt.title('Line Plot')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
x = np.arange(0,11)
y1 = x*2
y2 = y*3
plt.plot(x,y1,color='g', linewidth=2, linestyle=":")
plt.plot(x,y2,color='r', linewidth=2, linestyle="-.")
plt.title('Line Plot')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.grid(True)
plt.show()
```

```python
x = np.arange(0,11)
y1 = x*2
y2 = y*3
plt.subplot(1,2,1)                                             # row, column, order
plt.plot(x,y1,color='g', linewidth=2, linestyle=":")
plt.subplot(1,2,2)                                             # row, column, order
plt.plot(x,y2,color='r', linewidth=2, linestyle="-.")
plt.title('Line Plot')
plt.xlabel('X Label')
plt.ylabel('Y Label')
#plt.grid(True)


plt.show()
```

## Bar Plot

```python
student = {"Sam":67, "Matt":23, "Julia":98}
x = list(student.keys())
y = list(student.values())
plt.title('Bar Plot')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.bar(x,y,color='g')
plt.grid(True)
plt.show()
```

```python
student = {"Sam":67, "Matt":23, "Julia":98}
x = list(student.keys())
y = list(student.values())
plt.title('Bar Plot')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.barh(x,y,color='r')                                  #Horizontal Bar
plt.grid(True)
plt.show()
```

## Box Plot & Violin Plot

```python
x = np.random.randint(1,10,8)
y = np.random.randint(1,10,8)
z = np.random.randint(1,10,8)

data = [x,y,z]
plt.boxplot(data)
print(x, '\n',y,'\n',z,'\n')
plt.show()
```

```python
x = np.random.randint(1,10,8)
y = np.random.randint(1,10,8)
z = np.random.randint(1,10,8)

data = [x,y,z]
plt.violinplot(data)
print(x, '\n',y,'\n',z,'\n')
plt.show()
```

## Piechart & Doughnut Chart

```python
data_dict = {"Apple":34, "Guava":45,"Orange":23}
plt.pie(data_dict.values(), labels=data_dict.keys(), autopct='%0.1f%%', colors=['red','green','yellow'])
plt.show()
```

```python
data_dict = {"Apple":34, "Guava":45,"Orange":23}
plt.pie(data_dict.values(), labels=data_dict.keys(), autopct='%0.1f%%', colors=['red','green','yellow'], radius=2)  # outer pie with a bigger radius
plt.pie([1],colors=['white'], radius=1)   ## set 1 item with white clor and radius lesser than outer pie
plt.show()   
```
